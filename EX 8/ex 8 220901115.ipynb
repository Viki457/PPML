```python
import matplotlib.pyplot as plt
import numpy as np
ypoints= np.array ([3,8,1,10,5,7])
print(plt.plot(ypoints,linestyle='dashed'))
plt.title('sample graph')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.show()
```

```python
ypoints= np.array([3,8,1,10,5,7])
print(plt.plot(ypoints,marker='o'))
plt.show()
```
