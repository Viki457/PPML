```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D",'E'])
y=np.array([4,6,19,10,8])
plt.bar(x,y)
plt.show()
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="BWM",width=.5)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="Audi",color='r',width=.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance(kms)')
plt.title('information')
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```
